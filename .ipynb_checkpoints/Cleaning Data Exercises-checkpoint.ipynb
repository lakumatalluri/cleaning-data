## Missing Data

In the Towards Data Science article, they list 3 techniques for detecting any missing data frame. Try all three before deciding on a way to solve any missing data issues.

```python
# Plot a heatmap of any missing data
cols = df.columns[:30] # first 30 columns
colours = ['#000099', '#ffff00'] # specify the colours - yellow is missing. blue is not missing.
sns.heatmap(df[cols].isnull(), cmap=sns.color_palette(colours))
```

```python
# Try out a missing data percentage list! 
# if it's a larger dataset and the visualization takes too long can do this.
# % of missing.
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Plot a missing data histogram
# first create missing indicator for features with missing data
for col in df.columns:
    missing = df[col].isnull()
    num_missing = np.sum(missing)
    
    if num_missing > 0:  
        print('created missing indicator for: {}'.format(col))
        df['{}_ismissing'.format(col)] = missing


# then based on the indicator, plot the histogram of missing values
ismissing_cols = [col for col in df.columns if 'ismissing' in col]
df['num_missing'] = df[ismissing_cols].sum(axis=1)
df['num_missing'].value_counts().reset_index().sort_values(by='index').plot.bar(x='index', y='num_missing')
```

Now that you have tried all 3 techniques for detecting missing data, did you find any? Reading through the article, which solution do you think will help you clean the data set and handle the missing data? Try out your solution below!

```python
# Handle any missing data in your dataframe.
# If we want to drop feature Title.
cols_to_drop = ['Title']
df_dropTitle = df.drop(cols_to_drop, axis=1)
df_dropTitle.head()
```

```python
df_dropMissingValues=df_dropTitle.dropna(how='any',subset=['Review Text'],inplace=False)
df_dropMissingValues.shape
df_dropMissingValues.dropna(how='any',subset=['Division Name'],inplace=True)
print(df_dropMissingValues.shape)
df_dropMissingValues.dropna(how='any',subset=['Department Name'],inplace=True)
print(df_dropMissingValues.shape)
df_dropMissingValues.dropna(how='any',subset=['Class Name'],inplace=True)
print(df_dropMissingValues.shape)
#Check with heat map that there is no missing data left
cols = df_dropMissingValues.columns[:10] # first 30 columns
colours = ['#000099', '#FFFF00'] # specify the colours - yellow is missing. blue is not missing.
sns.heatmap(df_dropMissingValues[cols].isnull(), cmap=sns.color_palette(colours))
print(df_dropMissingValues.shape)
```

## Unnecessary Data

Unnecessary data could be duplicates, irrelevant, or any uninformative data. Dive into each type of unnecessary data to see if there is any in the data set. Make sure to make note of how you would handle each type of unnecessary data.

```python
# Look for any irrelevant data in the dataframe. How will you handle it?
num_rows = len(df_dropMissingValues.index)
low_information_cols = [] #
for col in df_dropMissingValues.columns:
    cnts = df_dropMissingValues[col].value_counts(dropna=False)
    top_pct = (cnts/num_rows).iloc[0]
    if top_pct > 0.95:
        low_information_cols.append(col)
        print('{0}: {1:.5f}%'.format(col, top_pct*100))
        print(cnts)
        print()
        #handling unnecessary data by removing the added columns during the missing data identification phase
df_dropUnnecessaryData = df_dropMissingValues.drop('Title_ismissing', axis=1)
df_dropUnnecessaryData = df_dropUnnecessaryData.drop('Review Text_ismissing', axis=1)
df_dropUnnecessaryData = df_dropUnnecessaryData.drop('Division Name_ismissing', axis=1)
df_dropUnnecessaryData = df_dropUnnecessaryData.drop('Department Name_ismissing', axis=1)
df_dropUnnecessaryData = df_dropUnnecessaryData.drop('Class Name_ismissing', axis=1)
df_dropUnnecessaryData = df_dropUnnecessaryData.drop('num_missing', axis=1)
print(df_dropUnnecessaryData.shape)
print(df_dropUnnecessaryData.columns)
```

```python
# Look for duplicates. How will you handle any duplicates?
# we know that column 'Unnamed: 0' is unique, but what if we drop it?
df_dedupped = df_dropUnnecessaryData.drop('Unnamed: 0', axis=1).drop_duplicates()
# there were duplicate rows
print(df_dropUnnecessaryData.shape)
print(df_dedupped.shape)
```
